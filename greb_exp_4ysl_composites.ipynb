### Copy the binary input files to the main directory

```python
import shutil as sh
import glob
import os
import re

main_dir = '/home/nicholat/project/greb/'
input_dir = main_dir + 'input_files/'
orig_input_dir = main_dir + 'orig_input/'

# copy the original input files into the main folder
for filename in glob.glob(os.path.join(orig_input_dir, '*')):
    sh.copy(filename, main_dir)
print "Original input files copied to main dir"
print "Anom placeholder bin files copied to main dir"

if exp == 'clim':
    print "Climate run, namelist changed"
    with open('namelist','r') as f_in:
        lines = f_in.readlines()
        # do the replacement in the ith line.
        lines[7] = re.sub('wnd','clm',lines[7])
        print lines[7]
    with open('namelist','w') as f_out:
        f_out.writelines(lines)
        
if exp == 'sst_anom.pos.4ycomp':
    sh.copyfile(input_dir+'sst.anom',
                main_dir+'sst.anom')
    print "Anomalous SST input copied to main dir"
    with open('namelist','r') as f_in:
        lines = f_in.readlines()
        # do the replacement in the ith line.
        lines[7] = re.sub('wnd','sst',lines[7])
        print lines[7]
    with open('namelist','w') as f_out:
        f_out.writelines(lines)


if exp == 'v_anom.pos.4ycomp':
    sh.copyfile(input_dir+'meridional.wind.4ycomp',
                main_dir+'meridional.wind.anom')
    print "Anomalous meridional wind input copied to main dir"
    
if exp == 'u_anom.pos.4ycomp':
    sh.copyfile(input_dir+'zonal.wind.4ycomp',
                main_dir+'zonal.wind.anom')
    print "Anomalous zonal wind input copied to main dir"
    
if exp == 'uv_anom.pos.4ycomp':
    sh.copyfile(input_dir+'zonal.wind.4ycomp',
                main_dir+'zonal.wind.anom')
    print "Anomalous zonal wind input copied to main dir"
    sh.copyfile(input_dir+'meridional.wind.4ycomp',
                main_dir+'meridional.wind.anom')
    print "Anomalous meridional wind input copied to main dir"

if exp == 'cld.pos.4ycomp':
    sh.copyfile(input_dir+'cloud.cover.cld.4ycomp',
                main_dir+'cloud.cover.anom')
    with open('namelist','r') as f_in:
        lines = f_in.readlines()
        # do the replacement in the ith line.
        lines[7] = re.sub('wnd','cld',lines[7])
        print lines[7]
    with open('namelist','w') as f_out:
        f_out.writelines(lines)

    print "Anomalous cloud clover input copied to main dir"
    
if exp == 'smc.pos.4ycomp':
    sh.copyfile(input_dir+'soil.moisture.smc.4ycomp',
                main_dir+'soil.moisture.anom')
    with open('namelist','r') as f_in:
        lines = f_in.readlines()
        # do the replacement in the ith line.
        lines[7] = re.sub('wnd','smc',lines[7])
        print lines[7]
    with open('namelist','w') as f_out:
        f_out.writelines(lines)

    print "Anomalous soil moisture input copied to main dir"

if exp == 'smc.neg.4ycomp':
    sh.copyfile(input_dir+'soil.moisture.smc.4ycomp',
                main_dir+'soil.moisture.anom')
    with open('namelist','r') as f_in:
        lines = f_in.readlines()
        # do the replacement in the ith line.
        lines[7] = re.sub('wnd','smc',lines[7])
        print lines[7]
    with open('namelist','w') as f_out:
        f_out.writelines(lines)

    print "Anomalous soil moisture input copied to main dir"
    
if exp == 'all.pos.4ycomp':
    sh.copyfile(input_dir+'zonal.wind.4ycomp',
                main_dir+'zonal.wind.anom')
    print "Anomalous zonal wind input copied to main dir"
    sh.copyfile(input_dir+'meridional.wind.4ycomp',
                main_dir+'meridional.wind.anom')
    print "Anomalous meridional wind input copied to main dir"
    sh.copyfile(input_dir+'cloud.cover.cld.4ycomp',
                main_dir+'cloud.cover.anom')
    print "Anomalous cloud clover input copied to main dir"
    sh.copyfile(input_dir+'soil.moisture.smc.4ycomp',
                main_dir+'soil.moisture.anom')
    print "Anomalous soil moisture input copied to main dir"
    
    with open('namelist','r') as f_in:
        lines = f_in.readlines()
        # do the replacement in the ith line.
        lines[7] = re.sub('wnd','all',lines[7])
        print lines[7]
    with open('namelist','w') as f_out:
        f_out.writelines(lines)
    print "Namelist changed to 'all' experiment"
    
if exp == 'all_sst.pos.4ycomp':
    sh.copyfile(input_dir+'zonal.wind.4ycomp',
                main_dir+'zonal.wind.anom')
    print "Anomalous zonal wind input copied to main dir"
    sh.copyfile(input_dir+'meridional.wind.4ycomp',
                main_dir+'meridional.wind.anom')
    print "Anomalous meridional wind input copied to main dir"
    sh.copyfile(input_dir+'cloud.cover.cld.4ycomp',
                main_dir+'cloud.cover.anom')
    print "Anomalous cloud clover input copied to main dir"
    sh.copyfile(input_dir+'soil.moisture.smc.4ycomp',
                main_dir+'soil.moisture.anom')
    print "Anomalous soil moisture input copied to main dir"
    sh.copyfile(input_dir+'sst.anom',
                main_dir+'sst.anom')
    print "Anomalous SST input copied to main dir"
    
    with open('namelist','r') as f_in:
        lines = f_in.readlines()
        # do the replacement in the ith line.
        lines[7] = re.sub('wnd','ast',lines[7])
        print lines[7]
    with open('namelist','w') as f_out:
        f_out.writelines(lines)
    print "Namelist changed to 'ast' experiment"

        
```
